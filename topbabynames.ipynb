```python
import pandas as pd
import numpy as np

df = pd.read_csv('https://public.tableau.com/app/sample-data/TopBabyNamesbyState.csv')
df.head(30)
df.tail(30)
df.info()
#no nulls
#count state, count year min max sort, unique top names
df.head(50)
df['Top Name'].nunique()
df['State'].nunique()
df['State'].value_counts()
#there are 51 states? DC
min(df['Year'])
max(df['Year'])
df.describe()

max(df['Occurences'])
df[df['Occurences'] == 10023]

df['Top Name'].nunique()
df['Top Name'].value_counts().head(50)
df[df['Top Name'] == 'Joe']

#check if there are duplicates, no duplicates detected
df.duplicated(keep=False).value_counts()
```

```python
#most popular name by state, all time, year
df.groupby(['Top Name'])['Occurences'].sum().sort_values(ascending=False)
#mary, michael, james, jennifer are some of the most popular names
df1 = df.groupby(['State','Top Name'])['Occurences'].sum().reset_index().sort_values(by=['State','Occurences'],ascending=[True,False])
df1.groupby(['State'])['Occurences'].idxmax()
df1.loc[df1.groupby(['State'])['Occurences'].idxmax()]
#most popular names by state
#divide by gender
df2 = df.groupby(['State','Gender','Top Name'])['Occurences'].sum().reset_index().sort_values(by=['State','Gender','Occurences'],ascending=[True,True,False])
df2.groupby(['State','Gender'])['Occurences'].idxmax()
df2.loc[df2.groupby(['State','Gender'])['Occurences'].idxmax()].head(50)
```
